Calcular o percentual de gordura pode não ser acessível a muitas pessoas, a avaliação física através de bioimpedância (de alta precisão) e pesagem hidrostática são avaliações caras. Você pode recorrer ao adipômetro para medir as dobras cutêneas, porém, manusear o equipamento e fazer as medições de forma correta não é tarefa fácil, e é recomendável que para maior precisão seja feito por um profissional da nutrição ou da educação física.

Porém, a Marinha Americana desenvolveu um método de avaliação de gordura corporal utilizando apenas uma **fita métrica**, material bastante acessível e fácil de manusear, será necessário apenas fazer a medição da circunferência de algumas partes do corpo e através de uma fórmula você poderá estimar qual a sua gordura corporal.

**A FÓRMULA FOI DESENVOLVIDA PARA PESSOAS ENTRE 20 E 79 ANOS**

Alguns códigos para a construção da calculadora:

```python
# Importando a funcao log10
from math import log10
```

```python
# Funcao para coletar as informacoes do usuario
def coletor_informacoes():
    
    sexo = input('Qual o seu sexo? H para Homem e M para mulher: ')
    idade = int(input('Qual a sua idade? '))
    altura = float(input('Digite sua altura (cm): '))
    cintura = float(input('Circurferência da cintura (cm): '))
    pescoco = float(input('Circurferência do pescoço (cm): '))

    info = {'sexo':sexo, 'idade':idade, 'altura':altura,
            'cintura':cintura, 
            'pescoco':pescoco}
    
    if sexo == 'M':
        quadril = float(input('Circunferência do quadril (cm): '))
        info['quadril'] = quadril
    
    return info
```

```python
# Funcao para calcular a gordura corporal de homens

def formula_para_homens(info):
    gordura_calculada = 495 / (1.0324 - 0.191 * log10(info['cintura'] - info['pescoco']) + 0.155 * log10(info['altura'])) - 450
    return round(gordura_calculada,2)
```

```python
# Funcao para calcular a gordura corporal de mulheres

def formula_para_mulheres(info):
    gordura_calculada = 495 / (1.296 - 0.350 * log10(info['quadril'] + info['cintura'] - info['pescoco']) + 0.221 * log10(info['altura'])) - 450
    return round(gordura_calculada,2)
```

```python
def resultado_homens(info, gordura_calculada):
        
    if info['idade'] >= 20 and info['idade'] <= 39:
        if gordura_calculada < 8:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        elif gordura_calculada >= 8 and gordura_calculada < 20:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        elif gordura_calculada >= 20 and gordura_calculada < 25:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Obesidade')
    
    if info['idade'] >= 40 and info['idade'] <= 59:
        if gordura_calculada < 11:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        elif gordura_calculada >= 11 and gordura_calculada < 22:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        elif gordura_calculada >= 22 and gordura_calculada < 28:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print('Obesidade')
    
    if info['idade'] >= 60 and info['idade'] <= 79:
        if gordura_calculada < 13:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        if gordura_calculada >= 11 and gordura_calculada < 22:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        if gordura_calculada >= 22 and gordura_calculada < 28:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Obesidade')
```

```python
def resultado_mulheres(info, gordura_calculada):
        
    if info['idade'] >= 20 and info['idade'] <= 39:
        if gordura_calculada < 21:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        elif gordura_calculada >= 21 and gordura_calculada < 33:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        elif gordura_calculada >= 33 and gordura_calculada < 39:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Obesidade')
    
    if info['idade'] >= 40 and info['idade'] <= 59:
        if gordura_calculada < 23:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        elif gordura_calculada >= 23 and gordura_calculada < 35:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        elif gordura_calculada >= 35 and gordura_calculada < 40:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Obesidade')
    
    if info['idade'] >= 60 and info['idade'] <= 79:
        if gordura_calculada < 24:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Pouca gordura')
        if gordura_calculada >= 24 and gordura_calculada < 36:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Saudável')
        if gordura_calculada >= 36 and gordura_calculada < 42:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Acima do peso')
        else:
            print(f'Você possui {gordura_calculada}% de gordura corporal, você se classifica como: Obesidade')
```

```python
def calculadora_gordura():
    info = coletor_informacoes()
    if info['sexo'] == 'H':
        gordura_calculada = formula_para_homens(info)
        resultado_homens(info, gordura_calculada)
    
    if info['sexo'] == 'M':
        gordura_calculada = formula_para_mulheres(info)
        resultado_mulheres(info, gordura_calculada)
```

**Calculadora pronta:**

```python
calculadora_gordura()
```
